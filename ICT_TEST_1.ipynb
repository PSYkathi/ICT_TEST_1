#1. Bank Account Manager
a. Simulate a bank account with deposit, withdrawal, and balance
enquiry operations

b. Use variables and data types to store account information

c. Implement conditional statements to check for sufficient balance
before withdrawal

d. Use functions to perform deposit, withdrawal, and balance inquiry
operations.


```python
def manager():
    balance = 0

    status = True
    while status:
        print('1. deposit')
        print('2. withdraw')
        print('3. balance')
        print('4. exit')

        user_input = input("Choose an option: ")

        if (user_input == '1'):
            deposit_amount = float(input("Enter the Deposit amount: "))
            balance = balance + deposit_amount
            print("Amount Deposited:", deposit_amount)
            print("the balance is:",balance)
        elif (user_input == '2'):
            withdraw = float(input("Enter the Withdrawel Amount: "))
            if withdraw <= balance:
                balance = balance - withdraw
                print("The amount Withdrawn", withdraw)
                print("the balance is:",balance)
            else:
                print("insufficient balance")
        elif (user_input == '3'):
            print("the balance is:",balance)
        elif (user_input == '4'):
            print('see ya')
            break
        else:
            print('invalid choice')
            break

manager()
```

#2. Weather Data Analyzer
a. Analyze weather data (temperature, humidity, etc.) for a given region

b. Use variables and data types to store weather data

c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)

d. Use functions to calculate average temperature and humidity, and print
results

```python
def weather():
  temp = []
  humid = []

  climat = True
  while climat:
    print("======================Know the Temperature========================")

    temperature = int(input("Enter the temperature = "))
    humidity = int(input("Enter the humidity = "))
    temp.append(temperature)
    humid.append(humidity)
    # cold = 20
    # normal = 21
    # humid = 28

    if (temperature < 20 and humidity<29):
      print("The report is: Cold")
      break
    elif (temperature <=20  and humidity<=29):
      print("The Report is: Normal")
      break
    elif (temperature >20 and humidity>29):
      print("The Report is: humid")
      break
    else:
      print("The number seems to be Fantasy")
      break
  avg_temp=sum(temp)/len(temp)
  avg_humid=sum(humid)/len(humid)
  print("avg_temperature:", avg_temp)
  print("ave_humid:", avg_humid)

weather()
```
